```python
response = requests.get('https://www.yellowpages.com/search?search_terms=coffee+shops&geo_location_terms=Philadelphia%2C+PA',headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
business_name = []
badges = []
for names in soup.findAll('div',{'class':'search-results organic'}):
    for name in names.findAll('a',{'class':'business-name'}):
        business_name.append(name.span.text.strip())
```

```python
for name in soup.findAll('div',{'class':'v-card','class': 'locality'}):
        badges.append(name.text.strip())
```

```python
badges
```

```python
yellowpagesdf = pd.DataFrame(
    {'business_name': business_name,
     'badges': badges,
    })
yellowpagesdf
```
